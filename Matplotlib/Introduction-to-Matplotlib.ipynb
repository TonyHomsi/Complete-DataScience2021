## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1 ,2, 3, 4]
y= [11, 22, 33, 44]

# 2. Setuplot
fig, ax = plt.subplots(figsize=(10,10)) # (width, heigh)

# 3. Plot data
ax.plot(x,y)


# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays

we want:
* Line plot
* Scatter plot
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use the same data using a Scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# You can make plots from a dictionary using Bar plot
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

```python
# Horizontal Bar
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options  for subplots

```python
# Subplot 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

# Plot to each diffrenet axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

# plot to each differnet index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
